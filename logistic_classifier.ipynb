# Encoding 'stage fear' and 'drained after socializing' and 'Personality'
### The main objective of this code is to prepare a DataFrame for analysis or modeling by:

* **Converting Binary Categories**: Changing 'Yes'/'No' string values in specific columns into numerical 1s and 0s.

* **Encoding Personality Types**: Transforming 'Extrovert' and 'Introvert' labels into numerical representations.

* **Streamlining the DataFrame**: Removing the original categorical columns after their numerical equivalents have been created, thus ensuring a cleaner and more efficient dataset.

```python
df["Stage_fear_encoded"]= df["Stage_fear"].map({"Yes": 1, "No": 0})
df["Drained_after_socializing_encoded"]= df["Drained_after_socializing"].map({"Yes": 1, "No": 0})
df.loc[:,"Personality"] = df["Personality"].map({"Extrovert": 1, "Introvert": 0})
df.drop(columns=["Stage_fear", "Drained_after_socializing"], inplace=True)
```

```python
df_input= df.drop(columns=["Personality"])
df_output= df["Personality"]
```

# Data Imputation
#### When evaluating imputation methods by examining the difference between the original and imputed Spearman correlation matrices, the goal is to find the method that introduces the least change. In this context, a heatmap with values closest to zero (appearing as very light or uniformly subtle colors) indicates superior performance, as it signifies that the imputation has best preserved the dataset's intrinsic correlation structure. Upon reviewing the provided corr_diff heatmaps, knn imputation with distance as weight penalty and using linear regression to predict missing values gives differences which are remarkably close to zero across all feature pairs. Conversely, knn imputation with uniform penalty and bayesian method show more pronounced deviations, suggesting their respective imputation techniques caused greater alterations to the original correlation patterns.


## BayesianRidge
## The main objective of this code is to:

* **Handle Missing Values**: Impute (fill in) missing data points in a DataFrame.

* **Utilize an Advanced Imputation Strategy**: Employ an iterative imputation method (also known as MICE - Multiple Imputation by Chained Equations) which models each feature with missing values as a function of other features and uses that estimate for imputation.

* **Prepare Data for Analysis**: Generate a new DataFrame (df_byr) where missing values have been estimated and filled, making the dataset ready for further analysis or machine learning model training.

```python
# Handling missing values using BayesianRidge
import matplotlib.pyplot as plt
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imputer_byr = IterativeImputer(max_iter=10, random_state=42)
df_byr = pd.DataFrame(imputer_byr.fit_transform(df_input), columns=df_input.columns)
plt.subplot(1,2,1)
df_byr["Time_spent_Alone"].plot(kind="kde")
plt.title('BayesianRidge Imputation')
plt.subplot(1,2,2)
df_input["Time_spent_Alone"].plot(kind="kde")
plt.title('Original Data')
plt.show()
# df_byr.corr(numeric_only=True,method="spearman").style.background_gradient(cmap='coolwarm')
corr_diff_byr=df_input.corr(numeric_only=True,method="spearman") - df_byr.corr(numeric_only=True,method="spearman")
corr_diff_byr.style.background_gradient(cmap='coolwarm')
```

```python
df_byr["Stage_fear_encoded"] = df_byr["Stage_fear_encoded"].abs().round().astype(int)
# df_byr["Stage_fear_encoded"].value_counts().plot(kind="bar", title="Stage Fear Distribution")
plt.subplot(1,2,1)
df_byr["Stage_fear_encoded"].value_counts().plot(kind="bar")
plt.title('BayesianRidge Imputation')
plt.subplot(1,2,2)
df_input["Stage_fear_encoded"].value_counts().plot(kind="bar")
plt.title('Original Data')
plt.show()
```

## KNNimputer with uniform weights

## KNNImputer with distance weights
## Handle Missing Values: Impute (fill in) missing data points (represented as NaN) within a DataFrame.

* **Utilize a Neighborhood-Based Strategy**: Employ the K-Nearest Neighbors (KNN) algorithm to estimate missing values, where the imputation is based on a weighted average of values from its nearest neighbors.

* **Specify Weighting Scheme**: Use "distance" weights, meaning closer neighbors have a greater influence on the imputed value than farther neighbors.

* **Generate Complete Data**: Produce a new DataFrame (df_knn_dst) where all missing values have been estimated and populated, making the dataset ready for further analysis or machine learning model training.

```python
from sklearn.impute import KNNImputer
imputer_knn = KNNImputer(n_neighbors=4, weights="distance")
df_knn_dst = pd.DataFrame(imputer_knn.fit_transform(df_input), columns=df_input.columns)
```

* **Visualize Transformed Distributions**: Generate and display distribution plots (histograms with KDE) for each transformed column to visually inspect the effect of the transformation.

# Method 1: Predict using KNN
## The main objectives of this code are to:
1. Prepare Data for Modeling: Select relevant features and split the dataset into training and testing sets.
2. Standardize Features: Scale the numerical features to ensure they contribute equally to the distance calculations in the KNN algorithm.
3. Train a Classification Model: Train a K-Nearest Neighbors classifier to predict 'Personality' (likely 'Extrovert' or 'Introvert' based on previous encoding).
4. Evaluate Model Performance: Assess the accuracy of the trained KNN model on unseen test data

```python
new_df.shape
```

```python
new_df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
x_features = new_df.drop(columns=["Personality","Going_outside","Stage_fear_encoded","Drained_after_socializing_encoded"])
y_target = new_df["Personality"]
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
# Knn applied with default parameters
knn = KNeighborsClassifier(n_neighbors=4, weights="uniform",p=2)
knn.fit(x_train_scaled, y_train)
y_pred = knn.predict(x_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy:.2f}")
```

```python
# Knn applied with distance weights and p=1
knn = KNeighborsClassifier(n_neighbors=4, weights="distance",p=1)
knn.fit(x_train_scaled, y_train)
y_pred = knn.predict(x_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy:.2f}")
```

```python
# Knn applied with distance weights and p=2
knn = KNeighborsClassifier(n_neighbors=4, weights="uniform",p=2)
knn.fit(x_train_scaled, y_train)
y_pred = knn.predict(x_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy:.2f}")
```

# Apply PCA on the input features
## The main objectives of this code are to:
1. Reduce Dimensionality by Retaining Key Variance: Apply PCA to the scaled features, specifically selecting a subset of principal components (eigenvectors) whose corresponding eigenvalues cumulatively capture at least 90% of the total variance (information) present in the original data. This reduces the number of dimensions while retaining most of the data's essential patterns.
2. Inspect Principal Components: Examine the learned principal components, which represent new, orthogonal features derived from the original ones.
3. Train a Classification Model on Reduced Data: Train a K-Nearest Neighbors classifier on these reduced-dimension features to predict 'Personality'.
4. Evaluate Model Performance: Assess the accuracy of the KNN model on the PCA-transformed test data.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.9)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)
pca.components_
```

## KNN Accuracy Remaining the Same:

1. K-Nearest Neighbors (KNN) is a distance-based algorithm. It calculates distances between data points in the feature space.
2. If the KNN model performs with similar accuracy on the original 7 scaled features as it does on the 3 PCA-transformed features, it indicates that the 3 principal components effectively represent all the "discriminative" information needed for the KNN classifier.
3. The redundant or highly correlated information in the original 7 features (which was largely discarded by reducing to 3 components) was not adding significant value or was even adding "noise" that the KNN algorithm could ignore, or that didn't fundamentally change the relative distances between data points in a way that impacted classification performance.

```python
# Knn applied with distance weights and p=2
knn = KNeighborsClassifier(n_neighbors=4, weights="uniform",p=2)
knn.fit(x_train_pca, y_train)
y_pred = knn.predict(x_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel().tolist()
print(f"True Positives: {tp}, True Negatives: {tn}, False Positives: {fp}, False Negatives: {fn}")
```
